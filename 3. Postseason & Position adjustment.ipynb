```python
import pandas as pd

# Uploading the stats file created earlier. Also uploading the playoff_data
stats = pd.read_csv("C:\\Users\\17667\\Desktop\\Python\\NYU Python course\\project\\NBA data\\stats\\stats.csv")
stats = stats.iloc[:,2:]
all_sheet = pd.read_excel("C:\\Users\\17667\\Desktop\\Python\\NYU Python course\\project\\NBA data\\info\\playoff_data.xlsx",sheet_name=None)
playoff = all_sheet['playoff']
semi_final = all_sheet['semi_final']
conf_final = all_sheet['conf_final']
final = all_sheet['final']
champ = all_sheet['champ']

# Next I want to add 5 columns to stats, which will keep record of the number of years of experience a player has.
new_columns = ['playoff', 'semi_final', 'conf_final', 'final', 'champ']
for col in new_columns:
    stats[col] = 0
# Now I will record the playoff, semi_final and all postseason appearances for each player in each season.
for i in range(stats.shape[0]):
    team = stats.iloc[i]['TEAM']
    season = stats.iloc[i]['season']
    if season <2024:
        if team in playoff[playoff['year']==season]['teams'].values[0]:
            stats.loc[i,'playoff'] += 1
        if team in semi_final[semi_final['year']==season]['teams'].values[0]:
            stats.loc[i,'semi_final'] += 1
        if team in conf_final[conf_final['year']==season]['teams'].values[0]:
            stats.loc[i,'conf_final'] += 1
        if team in final[final['year']==season]['teams'].values[0]:
            stats.loc[i,'final'] += 1
        if team in champ[champ['year']==season]['teams'].values[0]:
            stats.loc[i,'champ'] += 1


# Here I cumsum the appearances for each player to reflect their postseason experiences.
stats2 = stats

stats2 = stats2.sort_values(by =['PLAYER','season']).set_index(['PLAYER','season'])
stats2['playoff'] = stats2.groupby('PLAYER')['playoff'].cumsum()
stats2['semi_final'] = stats2.groupby('PLAYER')['semi_final'].cumsum()
stats2['conf_final'] = stats2.groupby('PLAYER')['conf_final'].cumsum()
stats2['final'] = stats2.groupby('PLAYER')['final'].cumsum()
stats2['champ'] = stats2.groupby('PLAYER')['champ'].cumsum()
stats2 = stats2.reset_index()
stats2.to_csv("C:\\Users\\17667\\Desktop\\Python\\NYU Python course\\project\\NBA data\\stats\\stats2.csv")
```
